```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('data/sample.csv').values
x = dataset[:,2]
x[6] = 55.0
y = dataset[:,3]
y[4] = 100000.0
```

```python
print(x)
print(y)
```

### 1. line plot

```python
plt.figure(figsize=(3,2))
# line plot: plt.plot(x), 데이터가 하나일 경우 가로축은 데이터의 idx
# plt.plot(x)
plt.plot(sorted(x), y)
```

```python
plt.figure(figsize=(7,4))
axis_x = np.arange(200, 1200, 100)
plt.plot(axis_x, x,
        color='green',
        marker='o',
        ms=10,
        linestyle='--',
        linewidth='3',
        label='chart1') # 한 figure에 여러 차트가 있을 때 각 차트의 이름을 정해줌

plt.plot(axis_x, x-10,
        color='red',
        marker='s',
        ms=10,
        linestyle=':',
        linewidth='3',
        label='chart2')
plt.legend() # label 표시
plt.title('line plot', fontsize=20, color='pink')
plt.xlabel('x')
plt.ylabel('age')
plt.grid() # 격자 표시
```

### 2. scatter plot

```python
plt.figure(figsize=(5,3))
plt.scatter(x, y,
            color='blue',
            marker='s',
            s=50,
            label='scatter 1')

plt.scatter(x-5, y+10000,
            color='red',
            marker='s',
            s=50,
            label='scatter 2')

plt.legend()
plt.title('scatter plot')
plt.grid()
plt.xlim([10,70])
plt.ylim([20000,150000])
# 범위 지정
```
